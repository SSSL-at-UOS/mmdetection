```python
config = '/home/soojin/mmdetection/project_work_dirs/multiple_damage_detection/configs/backbones/mask_rcnn_x101_64x4d_fpn_1x.py'
checkpoint = '/home/soojin/mmdetection/project_work_dirs/multiple_damage_detection/work_dirs/mask_rcnn_x101_64x4d_fpn_1x/epoch_42.pth'

cfg = mmcv.Config.fromfile(config)

# set cudnn_benchmark
if cfg.get('cudnn_benchmark', False):
    torch.backends.cudnn.benchmark = True
    
cfg.model.pretrained = None
cfg.data.test.test_mode = True

show = True 

# build the dataloader
# TODO: support multiple images per gpu (only minor changes are needed)
dataset = build_dataset(cfg.data.test)
data_loader = build_dataloader(
    dataset,
    imgs_per_gpu=1,
    workers_per_gpu=cfg.data.workers_per_gpu,
    dist=False,
    shuffle=False)

# build the model and load checkpoint
model = build_detector(cfg.model, train_cfg=None, test_cfg=cfg.test_cfg)
checkpoint = load_checkpoint(model, checkpoint, map_location='cpu')

# old versions did not save class info in checkpoints, this walkaround is
# for backward compatibility
if 'CLASSES' in checkpoint['meta']:
    model.CLASSES = checkpoint['meta']['CLASSES']
else:
    model.CLASSES = dataset.CLASSES
    
model = MMDataParallel(model, device_ids=[0])    
model.eval()
results = []
dataset = data_loader.dataset
prog_bar = mmcv.ProgressBar(len(dataset))

try : 
    for i, data in enumerate(data_loader):
    
        with torch.no_grad():
            result = model(return_loss=False, rescale=not show, **data)
        results.append(result)

        if show:
            model.module.show_result(data, result)

        batch_size = data['img'][0].size(0)
        for _ in range(batch_size):
            prog_bar.update()
except FileNotFoundError as  e: 
    print('Error') 
    
        
seg_results = []
num_class = len(dataset.CLASSES) + 1 # background + object classes 

for img_num, result in enumerate(results) : 
    
    bbox_result, segm_result = result
    
    for segm_result_ in segm_result : 
        if segm_result_:
            img_size = segm_result_[0]['size']
    seg_sum = np.zeros((img_size[0], img_size[1], num_class), dtype=np.float32)
    mask_sum = np.zeros((img_size[0], img_size[1]) , dtype=np.bool)

    for class_, bboxes in enumerate(bbox_result):
        
        class_idx = class_ + 1 

        for idx, bbox in enumerate(bboxes) : 
            confidence = bbox[-1]
            if confidence >= 0.5 : 
                mask = maskUtils.decode(segm_result[class_][idx]) 
                mask_sum = mask_sum  + mask
                seg_sum[:, :, class_idx] += confidence*mask

    seg_sum[:,:, 0] = mask_sum == 0

    seg_sum = torch.from_numpy(np.transpose(seg_sum, (2, 0, 1)))
    seg_sum = seg_sum.unsqueeze(0) # nn.softmax requires 4D tensor not 3D array 
    softmax = torch.nn.Softmax2d()
    output = softmax(seg_sum)
    _, indices = torch.max(output[0], 0)
    indices = indices.numpy()
    
    seg_results.append(indices)

```

```python
import skimage.io as io
import scipy
from skimage.measure import regionprops, label
import math


path = '/home/soojin/mmdetection/data/concrete_damage_autocon/test2018/'
im_list_path = glob.glob(os.path.join(path,'*.jpg')) + glob.glob(os.path.join(path,'*.JPG'))

# 49 47 48 
config = '/home/soojin/mmdetection/project_work_dirs/multiple_damage_detection/configs/bilin_112/mask_rcnn_r101_fpn_1x_mask_bilin_112_overlap.py'
checkpoint = '/home/soojin/mmdetection/project_work_dirs/multiple_damage_detection/work_dirs/mask_rcnn_r101_fpn_1x_mask_bilin_112_overlap/epoch_42.pth'
device = 'cuda:0'
model = init_detector(config, checkpoint, device=device)

annot_dir = os.path.join(path ,'annot')

average_precision = 0
average_recall = 0
average_precision_pix = 0
average_recall_pix = 0
for org_img in range(np.size(im_list_path)):    
# for org_img in range(10, 11):    
    
    image = io.imread(im_list_path [org_img])
    clear_output(wait=True)
    print(org_img, im_list_path[org_img])
    if image.ndim == 2 : image = color.gray2rgb(image)

    result = inference_detector(model, image)
    
    r={}
    r['class_ids'] = []
    r['rois'] = []
    r['scores'] = []
    r['masks'] = []

    bbox_result, segm_result = result


    for class_num, bbox_with_class in enumerate(bbox_result) : 
        for box_num, bbox in enumerate(bbox_with_class) :
            if bbox[-1] > 0.5 :
                r['class_ids'].append(class_num+1)
                r['rois'].append(bbox[:-1])
                r['scores'].append(bbox[-1])
                if len(r['masks']) == 0 :

                    r['masks'] = maskUtils.decode(segm_result[class_num][box_num]) 
                    r['masks'] = r['masks'][:, :, np.newaxis]
                    r['masks'] = np.asarray(r['masks'])
                else : 
                    r['masks'] = np.dstack([r['masks'], maskUtils.decode(segm_result[class_num][box_num])])


    r['class_ids'] = np.asarray(r['class_ids'] )
    r['rois'] = np.asarray(r['rois'], dtype = np.int64)
    r['scores']= np.asarray(r['scores'])
    cls = r['class_ids']


   
    img_name = os.path.splitext(os.path.basename(im_list_path[org_img]))[0]
    annot_list_path = glob.glob(os.path.join(annot_dir, img_name+'*.jpg'))
    
    annot_class_by_number = np.zeros((image.shape[0], image.shape[1], 4))
    
    precision_count = 0
    
#     kernel = np.ones((5,5),np.uint8)
    for annot in range(np.size(annot_list_path)):
        
        annot_image = io.imread(annot_list_path[annot])

        # 1 : crack 2:efflorescence 3: rebar 4: spalling
        annot_class_by_number[np.logical_and.reduce([annot_image[:, :, 0] > 200,
                                                     annot_image[:, :, 1] < 150,
                                                     annot_image[:, :, 2] < 150]), 0] = 255
#         annot_class_by_number[:,:,0] = cv2.dilate(annot_class_by_number[:,:,0],kernel,iterations = 1)
        
        
        annot_class_by_number[np.logical_and.reduce([annot_image[:, :, 0] < 150,
                                                     annot_image[:, :, 1] > 200,
                                                     annot_image[:, :, 2] < 150]), 1] = 255
#         annot_class_by_number[:,:,1] = cv2.dilate(annot_class_by_number[:,:,1],kernel,iterations = 1)
        
        annot_class_by_number[np.logical_and.reduce([annot_image[:, :, 0] < 150,
                                                     annot_image[:, :, 1] > 200,
                                                     annot_image[:, :, 2] > 200]), 2] = 255
#         annot_class_by_number[:,:,2] = cv2.dilate(annot_class_by_number[:,:,2],kernel,iterations = 1)
        
        annot_class_by_number[np.logical_and.reduce([annot_image[:, :, 0] > 200,
                                                     annot_image[:, :, 1] > 200,
                                                     annot_image[:, :, 2] < 150]), 3] = 255
#         annot_class_by_number[:,:,3] = cv2.dilate(annot_class_by_number[:,:,3],kernel,iterations = 1)

        
        labels, num = label(annot_class_by_number[:,:,1],connectivity=2, return_num= True)
    
        for R in regionprops(labels):
            if R.area > 1:
                annot_class_by_number[:,:,1]=scipy.ndimage.binary_fill_holes(annot_class_by_number[:,:,1]).astype('uint8')*255
                
        labels, num = label(annot_class_by_number[:,:,2],connectivity=2, return_num= True)
    
        for R in regionprops(labels):
            if R.area > 1:
                annot_class_by_number[:,:,2]=scipy.ndimage.binary_fill_holes(annot_class_by_number[:,:,2]).astype('uint8')*255
                
        labels, num = label(annot_class_by_number[:,:,3],connectivity=2, return_num= True)
    
        for R in regionprops(labels):
            if R.area > 1:
                annot_class_by_number[:,:,3]=scipy.ndimage.binary_fill_holes(annot_class_by_number[:,:,3]).astype('uint8')*255
    
    confusion_mat_img = np.zeros((image.shape[0], image.shape[1], 4)) 
    
    mask_by_class = np.zeros((image.shape[0], image.shape[1], 4))
    
    
    pix_level_detection = np.zeros((image.shape[0], image.shape[1], 4))
    confusion_matrix = np.zeros((image.shape[0], image.shape[1], 4))
    
    per_class_recall = np.zeros(4)
    per_class_precision = np.zeros(4)
    NTP_NFP_list = np.zeros(4)
    NTP_NFN_list = np.zeros(4)
    
    for i_cls in range(np.size(cls)): 
        pix_level_detection[r['masks'][:,:,i_cls]==1, r['class_ids'][i_cls]-1] = 255
    
    for i_cls in range(4): 
        confusion_matrix[:,:, i_cls] = pix_level_detection[:,:,i_cls] + annot_class_by_number[:,:,i_cls]
#         confusion_matrix[np.logical_and.reduce([pix_level_detection[:,:,i_cls] == 255, 
#                                                annot_class_temp[:,:,i_cls] == 0]), i_cls] = 2
#         confusion_matrix[np.logical_and.reduce([pix_level_detection[:,:,i_cls] == 0, 
#                                                annot_class_temp[:,:,i_cls] == 255]), i_cls] = 1
        
        NTP = np.sum(confusion_matrix[:,:,i_cls] > 500  )
        NTP_NFP = np.sum(pix_level_detection[:,:,i_cls] == 255)
        NTP_NFN = np.sum(annot_class_by_number[:,:,i_cls] == 255 )
        
        NTP_NFP_list = NTP_NFP
        NTP_NFN_list = NTP_NFN
#         print(NTP)
        
        per_class_recall[i_cls] = NTP / (NTP_NFN)
        per_class_precision[i_cls] = NTP / (NTP_NFP)
        
        
        
#     print(per_class_recall)    
#     print(per_class_precision) 
    pix_recall = np.nanmean(per_class_recall)
    pix_precision = np.nanmean(per_class_precision)
    

    
    average_precision_pix = average_precision_pix + pix_precision if not math.isnan(pix_precision) else average_precision_pix
    average_recall_pix = average_recall_pix + pix_recall  if not math.isnan(pix_precision) else average_recall_pix
        
    for i_cls in range(np.size(cls)): 
        confusion_mat = r['masks'][:,:,i_cls]+annot_class_by_number[:,:,r['class_ids'][i_cls]-1]/255
        mask_by_class[r['masks'][:,:,i_cls]==1, r['class_ids'][i_cls]-1] = 1
        try:
            pseudo_IoU = np.count_nonzero(confusion_mat == 2)/np.count_nonzero(r['masks'][:,:,i_cls] == 1)

        except ZeroDivisionError as e :
            pseudo_IoU = 0
        
#         print(pseudo_IoU)
        confusion_mat_img[r['masks'][:,:,i_cls], r['class_ids'][i_cls]-1] = 1  
        confusion_mat_img[confusion_mat == 2, r['class_ids'][i_cls]-1] = 2  
        if pseudo_IoU > 0.5 : 
            precision_count = precision_count + 1
            
    try :  
        precision = precision_count/np.size(cls)
        
    except ZeroDivisionError as e :
        precision = 0
    
    average_precision = average_precision + precision if not math.isnan(precision) else average_precision
    
    recall_count = 0 
    
    for annot in range(np.size(annot_list_path)):
        
        annot_image = io.imread(annot_list_path[annot])
        annot_class_temp = np.zeros((image.shape[0], image.shape[1], 4))
        
        
        
        # 1 : crack 2:efflorescence 3: rebar 4: spalling
        annot_class_temp[np.logical_and.reduce([annot_image[:, :, 0] > 200,
                                                     annot_image[:, :, 1] < 150,
                                                     annot_image[:, :, 2] < 150]), 0] = 255
#         annot_class_temp[:,:,0] = cv2.dilate(annot_class_temp[:,:,0],kernel,iterations = 1)
        
        annot_class_temp[np.logical_and.reduce([annot_image[:, :, 0] < 150,
                                                     annot_image[:, :, 1] > 200,
                                                     annot_image[:, :, 2] < 150]), 1] = 255
#         annot_class_temp[:,:,1] = cv2.dilate(annot_class_temp[:,:,1],kernel,iterations = 1)
        
        annot_class_temp[np.logical_and.reduce([annot_image[:, :, 0] < 150,
                                                     annot_image[:, :, 1] > 200,
                                                     annot_image[:, :, 2] > 200]), 2] = 255
#         annot_class_temp[:,:,2] = cv2.dilate(annot_class_temp[:,:,2],kernel,iterations = 1)
        
        annot_class_temp[np.logical_and.reduce([annot_image[:, :, 0] > 200,
                                                     annot_image[:, :, 1] > 200,
                                                     annot_image[:, :, 2] < 150]), 3] = 255
#         annot_class_temp[:,:,3] = cv2.dilate(annot_class_temp[:,:,3],kernel,iterations = 1)

        
        labels, num = label(annot_class_temp[:,:,1],connectivity=2, return_num= True)
        
        for R in regionprops(labels):
            if R.area > 1:
                annot_class_temp[:,:,1]=scipy.ndimage.binary_fill_holes(annot_class_temp[:,:,1]).astype('uint8')*255
                
        labels, num = label(annot_class_temp[:,:,2],connectivity=2, return_num= True)
        
        for R in regionprops(labels):
            if R.area > 1:
                annot_class_temp[:,:,2]=scipy.ndimage.binary_fill_holes(annot_class_temp[:,:,2]).astype('uint8')*255
                
        labels, num = label(annot_class_temp[:,:,3],connectivity=2, return_num= True)
    
        for R in regionprops(labels):
            if R.area > 1:
                annot_class_temp[:,:,3]=scipy.ndimage.binary_fill_holes(annot_class_temp[:,:,3]).astype('uint8')*255
        
        for i_cls in range(4): 
            if np.sum(annot_class_temp[:,:,i_cls]) > 0 :
                confusion_mat_recall = mask_by_class[:,:,i_cls]+annot_class_temp[:,:,i_cls]/255  
                try:
                    pseudo_IoU_recall = np.count_nonzero(confusion_mat_recall == 2)/np.count_nonzero(annot_class_temp[:,:,i_cls] == 255)

                except ZeroDivisionError as e :
                    pseudo_IoU_recall = 0
                
                if pseudo_IoU_recall > 0.5 : 
                    recall_count = recall_count + 1

    try :  
        recall = recall_count/np.size(annot_list_path)
        
    except ZeroDivisionError as e :
        recall = 0
    
    average_recall = average_recall + recall if not math.isnan(recall) else  average_recall  
                
average_recall = average_recall/np.size(im_list_path)
average_precision = average_precision/np.size(im_list_path)
average_recall_pix = average_recall_pix/np.size(im_list_path)
average_precision_pix = average_precision_pix/np.size(im_list_path)

print(average_precision)
print(average_recall)
print(average_precision_pix)
print(average_recall_pix)
        
```
